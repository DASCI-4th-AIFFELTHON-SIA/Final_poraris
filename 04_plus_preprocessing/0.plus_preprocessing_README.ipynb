# 전처리 목표 - 최종 정제 단계
추가적인 전처리로 노이즈를 줄이고 핵심 기사를 더 안정적으로 활용하기

## 문제점

기존 필터링을 통해 북한 관련 기사를 선별하였으나, 여전히 분석에 방해되는 데이터가 존재한다.    
이에 따라, 아래와 같은 추가 전처리를 수행하였습니다.

### 해결점

1. 제목에 ‘북한’ 또는 ‘北’이 포함된 기사만 추출
    * 기존 키워드 필터만으로는 일부 비관련 기사도 포함되는 문제
    * 제목에 ‘북한’ 또는 ‘北’이라는 단어가 명시적으로 포함된 기사만을 선별
    * 핵심 주제와 무관한 잡음 데이터를 최소화

2. 날씨 관련 기사 제거
    * 날씨 기사는 지명이 다수 포함되어 있지만 실제 사건이나 이슈와 관련이 없다.
    * 이러한 기사는 위치 정보 추출 시 잘못된 장소를 인식하게 만드는 주요 노이즈가 됨
    * 정확한 위치 정보 추출을 위해 불필요한 콘텐츠 제거

3. 중복된 기사 제거
    * 크롤링 과정에서 동일한 기사가 여러 번 수집되는 현상이 있었따.
    * 같이 기사가 날짜, 카테고리 등에 따라 여러번 노출된 경우 발생한 현상이다.
    * id, title, url 등을 기준으로 중복 여부를 확인하여 중복 데이터를 제거
    * 데이터의 정확성과 효율성을 높히기 위해 중복을 제거
