# 전처리 목표
10년치 연합뉴스 크롤링 원본 중 북한 관련 핵심 기사만을 안정적으로 수집한다.

## 문제점 
    * 10년치 뉴스기사는 양이 너무 많고 핵심 내용이 너무 다양하다.
    * 너무 기준을 좁게 잡으면 중요한 기사를 놓치고, 너무 넓게 잡으면 노이즈가 많아진다.
    * 본문이 짧은 기사들은 의미 있는 정보가 거의 없다.

---

## 해결점

* 1차 전처리
    * 이슈를 놓치지 않게 많은 키워드를 사용했다.
    * 이 키워드들이 제목 또는 본문에 있으면 해당 기사를 수집함.

* 2차 전처리
    * 본문 길이가 일정 기준보다 짧은 기사들은 제외

* 3차 전처리
    * 방송 안내형이나 특정 키워드의 기사는 의미 있는 정보가 없는것을 확인했다.
    * 특정한 제목 패턴을 가진 기사들을 한 번 더 제거
